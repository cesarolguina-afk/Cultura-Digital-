# Sprays. - Aromaterapia

### Que hacemos?
*Hacemos un spray principalmente usado para aromaterapia debido a las plantas medicinales y sus propiedades las cuales ayudan a multiples satisfacciones del olfato y del ambiente.
Nuestro producto contiene tres hierbas principales, Ruda, Romero y Albahcar. Su proceso no es complejo, para comenzar su elaboracion, necesitamos que las hierbas se encuentren frescas para que cuando depositemos el alcohol, sus propiedades sean mayormente aprovechadas.*




### Qu ingredientes usamos y por qu?

1.-*RUDA*.-*La utilizamos porque tiene propiedades
calmates
y antiflamatorias, ayuda a disminuir el estrs y su aroma
genera una sensacin de tranquilidad*.

2.-*ROMERO.-Reduce la tencin, el estrs,
 es muy bueno
para despejar la mente y equilibra el ambiente*.

3.-*ALBAHCAR.-Tiene un aroma fresco
y suave que ayuda a
disminuir la ansiedad, se usa mucho en sprays
para promover relajacin emocional*.

4.- *PASTILLAS DE ALCANFOR.-Tiene propiedades
aromaticas
similares al mentol usadas principlmente para descongestionar
y crear sensacion refrescante*.



### Procedimiento
 Las hierbas son agregadas al recipiente sin ramas ni impurezas para evitar la contaminacion del spray y se aprovechen todas las propiedades de las hierbas, por cada litro de alcohol, se agregan 100 gramos de cada hierba y tres pastillas de alcanfor, la cual nos ayuda a que las hierbas suelten sus propiedades de manera mas rapida, tome un tono verdoso y sea agradable al olfato reduciendo el olor del alcohol presente.
 Posteriormente, dejamos reposar por treinta dias para que se concentre el producto.

### Por qu elegimos el producto?
*Elegimos este prodcuto por el querer hacer que las personas tengan mayor conocimiento de los beneficios de las plantas y usados en aromaterapia, asi como los beneficios de la misma, teniendo mejores resultados que otro tipo de medicinas qumicas-farmacuticas.
Buscamos que las personas tengan acceso a este tipo de terapias con beneficios como control de estrs, descongestionar, calmar y mejorar el ambiente de lugares e incluso es usada en terapias holsticas y espirituales.*
