### Ejercicio 1
Pide la hora (0-23). Si es menor que 12 muestra 'Es mañana', si no 'Es tarde o noche'.


```python
# Escribe tu código aquí
var1=int(input("Dame la hora"))
if(var1>12):
 print("Es de tarde")
elif(var1<12):
 print("Es de mañana")
else:
  print("Son las 12")
```

### Ejercicio 2
Pide dos números y muestra cuál es mayor o si son iguales.


```python
# Escribe tu código aquí
var1=int(input("Dame un numero"))
var2=int(input("Dame otro numero"))
if(var1>var2):
 print("Es menor que")
elif(var1<var2):
 print("Es menor que")
else:
  print("son iguales")
```
